```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from SOM_plus_clustering.modules.som import SOM
import time
```

```python
df = pd.read_csv("iris_data.csv", header=None).iloc[:,:-1]
X = df.values
```

```python
method_list = ["random", "kde", "kmeans", "kde_kmeans", "kmeans++", "SOM++"]
total_trial = 100
df = pd.read_csv("Datas/time_data.csv")
data_table = list(df.values)
for trial in range(total_trial):
    time_trial = []
    for method in method_list:
        starttime = time.time()
        model = SOM(2,2,X.shape[1],initiate_method=method, learning_rate=1, neighbour_rad=1)
        pred = model.fit_predict(X=X, epoch=1, verbose=False)
        time_trial.append(time.time()-starttime)
    print(trial, "completed in", sum(time_trial))
    data_table.append(time_trial)
    df = pd.DataFrame(data_table, columns=method_list)
    df.to_csv("Datas/time_data.csv", index=False)
```
